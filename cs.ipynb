## Credit Score Classification Model
One more important feature (Credit Mix) in the dataset is valuable for determining credit scores. The credit mix feature tells about the types of credits and loans you have taken.

As the Credit_Mix column is categorical, I will transform it into a numerical feature so that we can use it to train a Machine Learning model for the task of credit score classification:

```python
data["Credit_Mix"] = data["Credit_Mix"].map({"Standard": 1, 
                               "Good": 2, 
                               "Bad": 0})
```

Now I will split the data into features and labels by selecting the features we found important for our model:

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming 'data' is your DataFrame
x = np.array(data[["Annual_Income", "Monthly_Inhand_Salary", 
                   "Num_Bank_Accounts", "Num_Credit_Card", 
                   "Interest_Rate", "Num_of_Loan", 
                   "Delay_from_due_date", "Num_of_Delayed_Payment", 
                   "Credit_Mix", "Outstanding_Debt", 
                   "Credit_History_Age", "Monthly_Balance"]])

y = np.array(data[["Credit_Score"]])

# Splitting data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.33, random_state=42)

# List of models to evaluate
models = [
    ("Logistic Regression", LogisticRegression(max_iter=1000)),
    ("Support Vector Machine", SVC()),
    ("Decision Tree", DecisionTreeClassifier()),
    ("Random Forest", RandomForestClassifier())
]

# Loop through each model, fit it, and evaluate
for name, model in models:
    model.fit(xtrain, ytrain.ravel())  # Fit model
    ypred = model.predict(xtest)  # Predict
    print(f"{name} Accuracy:", accuracy_score(ytest, ypred))
    print(classification_report(ytest, ypred))
```

## 1. Model Performances:
### Logistic Regression:
Accuracy: ~59%
The model is struggling to converge, as indicated by the ConvergenceWarning.
Precision and recall scores for each category (Good, Poor, Standard) show that this model may not be the best choice for our data.
### Support Vector Machine (SVM):
Accuracy: ~53%
Low recall for Good and Poor classes indicates that the model is biased towards the Standard class.
### Decision Tree:
Accuracy: ~75%
Relatively good precision, recall, and F1-scores, indicating that it performs well but may slightly overfit.
### Random Forest:
Accuracy: ~81%
The best performance among the evaluated models.
Balanced precision, recall, and F1-scores across all classes.
